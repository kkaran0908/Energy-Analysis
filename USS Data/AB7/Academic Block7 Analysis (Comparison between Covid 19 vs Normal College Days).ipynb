```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d1 = pd.read_excel('9MARCH_H.xlsx')
d2 = pd.read_excel('24FEB_H.xlsx')
```

```python
d1.std()
```

```python
d2.std()
```

## comparison between phase1 only

```python
d1['AC1_B7_SUBSTN_LT_EM1_PH1_VOLT.PresentValue'].plot()
d2['AC1_B7_SUBSTN_LT_EM1_PH1_VOLT.PresentValue'].plot()
#d1['AC1_B7_SUBSTN_LT_EM1_PH3_VOLT.PresentValue'].plot()
plt.title("Present Value of Voltage in Each Phase")
plt.legend(["Phase1 during Covid 19","Phase1 During Normal Time"])
plt.ylabel("Voltage",size = 13)
plt.xlabel("Time")
plt.plot()
```

```python
d1['AC1_B7_SUBSTN_LT_EM1_PH1_CURR.PresentValue'].plot()
d2['AC1_B7_SUBSTN_LT_EM1_PH1_CURR.PresentValue'].plot()  #curent in the phase 2 is 0, but voltage is not zero one of the reason might be the reading problem
#d1['AC1_B7_SUBSTN_LT_EM1_PH3_CURR.PresentValue'].plot()
plt.title("Present Value of Current in Each Phase")
plt.legend(["Phase1 during Covid 19","Phase1 During Normal Time"])
plt.ylabel("Current",size = 13)
plt.xlabel("Time")
plt.plot()
```

```python
d1['AC1_B7_SUBSTN_LT_EM1_PWR_R.PresentValue'].plot()
d2['AC1_B7_SUBSTN_LT_EM1_PWR_R.PresentValue'].plot()
plt.title("Power in Phase1")
plt.legend(["Phase1 during Covid 19","Phase1 During Normal Time"])
plt.ylabel("Power",size = 13)
plt.xlabel("Time")
plt.plot()
```

### Bar Plot for standard deviation Comparison

```python
import matplotlib.pyplot as plt

X = np.arange(22)
plt.bar(X + 0.00, d1.std().values, color = 'b', width = 0.25)
plt.bar(X + 0.25, d2.std().values, color = 'g', width = 0.25)
#plt.bar(X + 0.50, data[2], color = 'r', width = 0.25)

plt.show()
```
